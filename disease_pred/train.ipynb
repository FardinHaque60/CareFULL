```python
import pandas as pd
import xgboost as xgb
import sklearn
from sklearn.preprocessing import LabelEncoder
pd.options.display.max_columns = 4000
```

```python
train_data_path = "./data/Training.csv"
df = pd.read_csv(train_data_path)
df.head()
```

```python
df.describe()
df.shape
```

```python
for col in df.columns:
    print(col)
```

```python
unique = df.drop_duplicates()
unique
```

```python
unique.shape, unique['prognosis'].value_counts().count(), df['prognosis'].value_counts().count()
```

```python
data = unique.drop(["prognosis", "Unnamed: 133"], axis=1)
data
```

```python
labels = unique["prognosis"]
labels
```

```python
encoder = LabelEncoder()
```

```python
encoder.fit(labels)
enc_labels = encoder.transform(labels)
```

```python
clf = xgb.XGBClassifier()
```

```python
clf.fit(data, enc_labels)
```

```python
preds = clf.predict(data)
```

```python
for x, y in zip(enc_labels, preds):
    print(x, y)
```

```python
test_data_path = "./data/Testing.csv"
test_df = pd.read_csv(test_data_path)
test_df.head()
```

```python
test_data = test_df.drop(["prognosis"], axis=1)
test_data
```

```python
test_labels = test_df["prognosis"]
test_labels
```

```python
test_preds = clf.predict(test_data)
test_preds
```

```python
print("truth pred")
for x, y in zip(encoder.transform(test_labels), test_preds):
    print(x, y)
```
